```python
# Statcast for a certain time period for an example pitcher
lester_statcast = pb.statcast_pitcher(start_dt='2021-06-24', end_dt='2021-09-27', player_id=jlester['key_mlbam'].values[0])

# Filter for non-deprecated columns
pattern = re.compile('.*deprecated.*')
non_deprecated = [col for col in lester_statcast.columns if not re.match(pattern, col)]
lester_statcast = lester_statcast[non_deprecated]
lester_statcast[non_deprecated].columns
lester_statcast.head(4)
```

```python
# THE WRONG WAY TO GET PITCHER PLUS BATTER INFO. SHOW THIS ON THE STREAM TO SHOW PEOPLE A DANGEROUS THOUGHT PROCESS.
df = pd.DataFrame(lester_statcast[non_deprecated].groupby(['pitcher', 'batter']).batter.count()).reindex(axis=1)
sorted_lester_vs_batters = df.rename(columns={'batter':'Count'}).reset_index().sort_values(by='Count', ascending=False)
sorted_lester_vs_batters
```

```python
# Pitchers vs a batter
df = pd.DataFrame(lester_statcast[non_deprecated].groupby(['pitcher', 'batter', 'game_date']).at_bat_number.nunique()).reset_index()
sorted_lester_vs_batters = df.groupby(['pitcher', 'batter']).at_bat_number.sum().reset_index()
sorted_lester_vs_batters
```

```python
# Different Event Counts
event_counts = pd.DataFrame(lester_statcast[non_deprecated].groupby(['pitcher', 'batter', 'events']).events.count()).reindex(axis=1)
events = event_counts.rename(columns={'events':'Count'}).reset_index()
events
```

```python
events.events.unique()
```

```python
# Merge the At-Bats DF and the Events DF Together to get a Pct of the time
merged = events.merge(sorted_lester_vs_batters, on = ['pitcher', 'batter'], how='inner')
merged.rename(columns = {'at_bat_number':'Total_ABs'}, inplace=True)
merged['Pct'] = merged['Count'] / merged['Total_ABs']

# How many times has he struk out these hitters?
strikeouts = merged.loc[merged['events'] == 'strikeout'].groupby(['pitcher']).agg({'Count':'sum'}).values

# Out of how many at bats?
unique_abs = merged.Total_ABs.sum()
print(f'Out of {int(unique_abs)} at-bats, Jon Lester recorded {int(strikeouts)} strikeouts. A K-Rate of {round(float(strikeouts/unique_abs),3)}%.')
```
