```python
kershaw_row = playerid_lookup('kershaw', 'clayton')
kershaw_keymlbam = kershaw_row[['key_mlbam']].values[0][0]
```

```python
# His MLBAM ID is 477132, so we feed that as the player_id argument to the following function
from pybaseball import statcast_pitcher
kershaw_stats = statcast_pitcher('2017-06-01', '2017-07-01', kershaw_keymlbam)
kershaw_stats.head(2)
```

```python
# Filter for non-deprecated columns
pattern = re.compile('.*deprecated.*')
non_deprecated = [col for col in kershaw_stats.columns if not re.match(pattern, col)]
kershaw_stats = kershaw_stats[non_deprecated]
kershaw_stats.head(4)
```
